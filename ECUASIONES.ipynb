```python
import matplotlib.pyplot as plt
import numpy as np

# Ejercicio 1a: 5x - 2y = 1, 3x + 3y = 5
x = np.linspace(-3, 3, 100)

# Ecuación 1: y = (5x - 1)/2
y1 = (5*x - 1) / 2

# Ecuación 2: y = (5 - 3x)/3
y2 = (5 - 3*x) / 3

# Gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, y1, 'b-', linewidth=2, label='5x - 2y = 1')
plt.plot(x, y2, 'r-', linewidth=2, label='3x + 3y = 5')

# Solución: (1/3, 4/3)
plt.plot(1/3, 4/3, 'go', markersize=8, label='Solución (0.33, 1.33)')

plt.grid(True)
plt.xlim(-2, 2)
plt.ylim(0, 3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ejercicio 1a')
plt.legend()
plt.show()
```

```python
# Importa las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt


soluciones_verificadas = []
pares_a_verificar = [(0, 10), (1.5, 19), (-1, -4), (0, 0.4), (-0.5, 7)]

print("Verificando las soluciones para -3x + (1/2)y = 5:")
for x_val, y_val in pares_a_verificar:
    resultado = -3 * x_val + 0.5 * y_val
    if abs(resultado - 5) < 1e-9: # Usamos una tolerancia para la precisión de punto flotante
        print(f"El par ({x_val}, {y_val}) ES una solución.")
        soluciones_verificadas.append((x_val, y_val))
    else:
        print(f"El par ({x_val}, {y_val}) NO ES una solución.")

print("\n")

# b) Representar gráficamente la recta
# Para graficar, despejamos 'y': y = 6x + 10
print("Gráfico de la recta:")

x = np.linspace(-3, 3, 100)
y = 6 * x + 10

plt.figure(figsize=(7, 5))
plt.plot(x, y, label='-3x + (1/2)y = 5')

# Marcar las soluciones correctas en el gráfico
colores = ['go', 'ro', 'bo']
for i, (x_sol, y_sol) in enumerate(soluciones_verificadas):
    plt.plot(x_sol, y_sol, colores[i], markersize=8, label=f'Solución {i+1}: ({x_sol}, {y_sol})')

plt.title('Gráfico de la recta y sus soluciones')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()

# c) Relación entre los puntos de la recta y las soluciones
# La línea es la representación de todas las infinitas soluciones de la ecuación.
```
